# Approximating PI with the Monte-Carlo-Method

```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 20000000 #Number of random points

#Creating two array with N uniformly distibuted values between 0 and 1
x = np.random.uniform(size=N)
y = np.random.uniform(size=N)

```

```python
#Making Scatter plot of the random points
#Using maximal 1000 points
x_for_plot = x[0 : min(1000,N)]
y_for_plot = y[0 : min(1000,N)]
plt.scatter(x_for_plot, y_for_plot)

#Drawing the filled unitcircle
c1 = plt.Circle((0,0),1,color='r',alpha= 0.1)
plt.gcf().gca().add_artist(c1)
plt.axis([0,1,0,1])

#Plotting the unit circle (blue)
X  = [i / 1000 for i in range(0,1000)]
Y = [np.sqrt(1-(x)**2) for x in X]
plt.plot(X,Y)
```

```python
#Creating a matrix qith the distance from the center of the unit circle for each point
r = x*x + y*y 
inside_unit_circle = np.count_nonzero(r<1)
outside_unit_circle = N - inside_unit_circle
print('{} points are inside the unit circle coresponding to {}% of all points.'.format(
    inside_unit_circle,
    inside_unit_circle/N*100))
```

```python
#Approximation Pi with the Monte Carlo methond
pi_mc_approx = 4 * inside_unit_circle/N
print('With the monte carlo method pi was approximated to {:0.8f}.'.format(pi_mc_approx))
print('Pi is exactly: {:0.8f}. \nSo our apporximation is off by {:0.8f}.'.format(np.pi, pi_mc_approx -np.pi))
```
